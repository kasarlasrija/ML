*Hypotheses:*  T-TEST
- H0: distributionof tip for yellow = distribution of tip for green
- H1: distributionof tip for yellow â‰  distribution of tip for green

```python
yellow_tip=df['tip'][df.color=='yellow']
green_tip=df['tip'][df.color=='green']
```

```python
green_tip.shape
```

```python
yellow_tip.shape
```

```python
# Independent T-Test
from scipy import stats
t_stat, p_val = stats.ttest_ind(yellow_tip, green_tip)
print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
alpha=0.05 #Significance level
if p_val < alpha:
    print("Reject null hypothesis: Distribution of tip for yellow â‰  distribution of tip for green")
else:
    print("Fail to reject null hypothesis: Distribution of tip for yellow = distribution of tip for green")
```

```python
from scipy import stats

# Hypotheses:
# H0: Distribution of tip with credit card = distribution of tip with cash
# H1: Distribution of tip with credit card â‰  distribution of tip with cash

# Split tip data based on payment type
credit_s = df['tip'].loc[df['payment'] == 'credit card']
cash_s = df['tip'].loc[df['payment'] == 'cash']
```

```python
# Check sample sizes
print("Credit card samples:", credit_s.shape)
print("Cash samples:", cash_s.shape)

# Independent two-sample t-test
t_stat, p_val = stats.ttest_ind(credit_s, cash_s)

print("Independent T-Test statistic:", t_stat)
print("P-value:", p_val)
```

```python
# Significance level
alpha = 0.05
# Hypothesis decision
if p_val < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
# Formulate Hypotheses for ANOVA:
# H0: The mean tip amounts are the same across all 'dropoff_borough' categories.
# H1: At least one 'dropoff_borough' category has a different mean tip amount.
```

```python
unique_dropoff_boroughs = df['dropoff_borough'].dropna().unique()
dropoff_borough_tips = {}
for borough in unique_dropoff_boroughs:
    dropoff_borough_tips[borough] = df['tip'].loc[df['dropoff_borough'] == borough]
# Print the shapes of the extracted tip series to verify
for borough, tips_series in dropoff_borough_tips.items():
    print(f"'{borough}' tips shape: {tips_series.shape}")
```

```python
from scipy import stats
# Prepare the data for ANOVA from the dictionary
borough_tip_data = [tips_series for tips_series in dropoff_borough_tips.values()]
# Perform ANOVA test
f_stat_dropoff, p_val_dropoff = stats.f_oneway(*borough_tip_data)
print("ANOVA F-statistic for dropoff_borough:", f_stat_dropoff)
print("P-value for dropoff_borough:", p_val_dropoff)
alpha = 0.05 # Significance level
if p_val_dropoff < alpha:
    print("Reject null hypothesis: At least one 'dropoff_borough' category has a different mean tip amount.")
else:
    print("Fail to reject null hypothesis: The mean tip amounts are the same across all 'dropoff_borough' categories.")
```

```python
from scipy import stats

# H0: Distribution of tips is the same across all pickup_zone categories
# H1: At least one pickup_zone category has a different distribution of tips

# Prepare data
categories = [
    group['tip'].values
    for name, group in df.groupby('pickup_zone')
]

# Perform ANOVA
f_stat, p_val = stats.f_oneway(*categories)

print("ANOVA F-statistic:", f_stat)
print("P-value:", p_val)

# Decision rule
if p_val < 0.05:
    print("Reject null hypothesis: tips differs by pickup_zone")
else:
    print("Fail to reject null hypothesis: No significant difference by pickup_zone")
```

```python
from scipy import stats
# Hypotheses
# H0: Mean tip is the same across all pickup zones
# H1: At least one pickup zone has a different mean tip
feb_tip = df.loc[df['pickup_month'] == 'February', 'tip']
march_tip = df.loc[df['pickup_month'] == 'March', 'tip']
april_tip = df.loc[df['pickup_month'] == 'April', 'tip']

from scipy.stats import f_oneway
f_stat, p_value = f_oneway(feb_tip, march_tip, april_tip)
print("Pickup Month - F-statistic:", f_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject H0: There is a significant difference in tips by pickup month")
else:
    print("Accept H0: No significant difference in tips by pickup month")
```
